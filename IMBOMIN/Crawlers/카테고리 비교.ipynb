```python
# Ignore the warnings
import warnings
# warnings.filterwarnings('always')
warnings.filterwarnings('ignore')

# System related and data input controls
import os
import pandas as pd
pd.set_option('display.max_colwidth', None)  # 열 너비 설정을 None으로 지정하여 전체 URL을 표시
import numpy as np

df1 = pd.read_csv(r'D:\Analysis\DEBA\navernews_jan22.csv')
df2 = pd.read_csv(r'D:\Analysis\DEBA\navernews_jan22_new.csv')
df_mar1 = pd.read_csv(r'D:\Analysis\DEBA\navernews_mar22.csv')
df_mar2 = pd.read_csv(r'D:\Analysis\DEBA\navernews_mar22_new.csv')
```

```python
# 중복된 행을 추출하고 출력하는 함수
def find_duplicate_rows(df):
    duplicate_rows = df[df.duplicated(keep=False)]
    print(duplicate_rows)

# 함수 호출하여 중복된 행 출력
print("중복행 in df1:")
print(find_duplicate_rows(df1))
print("중복행 in df2:")
print(find_duplicate_rows(df2))
print("중복행 in df_mar1:")
print(find_duplicate_rows(df_mar1))
print("중복행 in df_mar2:")
print(find_duplicate_rows(df_mar2))
```

```python
# 중복 행 갯수 세는 함수
def find_duplicate_row_counts(dataframes):
    results = []
    
    for dataframe in dataframes:
        # 중복된 행의 수 계산
        duplicate_row_count = dataframe[dataframe.duplicated(keep=False)].shape[0]
        
        results.append(duplicate_row_count)
    
    return results

# 함수 호출 및 결과 출력
dataframes = [df1, df2, df_mar1, df_mar2]
duplicate_row_counts = find_duplicate_row_counts(dataframes)

for i, count in enumerate(duplicate_row_counts):
    print(f"DataFrame {i+1}의 중복된 행의 갯수:", count)
```
