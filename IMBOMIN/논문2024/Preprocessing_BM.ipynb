```python
# Ignore the warnings
import warnings
# warnings.filterwarnings('always')
warnings.filterwarnings('ignore')

import os
import pandas as pd

# 데이터 폴더 경로 설정
# 우선 알리익스프레스에 대하여 예시 진행
data_folder = r'C:\Users\Ellie\Analysis\Research\Data\알리익스프레스'

# 폴더 내 모든 CSV 파일 리스트 가져오기
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]

# 모든 CSV 파일을 데이터프레임으로 읽어 리스트에 저장
data_frames = []
for file in csv_files:
    file_path = os.path.join(data_folder, file)
    df = pd.read_csv(file_path)
    data_frames.append(df)
```

```python
# 데이터프레임 병합
merged_df = pd.concat(data_frames, ignore_index=True)
merged_df
```

```python
#불필요한 목록 제외한 df 생성
# 우선 날짜, 제목만 사용
news_df = merged_df[['Date', 'Title']]
#결측값 제거
news_df.dropna(subset=['Date', 'Title'], inplace=True)
news_df
```

```python
# 제거할 키워드 리스트(축구관련+가수 알리)
delete_keywords = ['토트넘', '손흥민', '축구', '스포츠', '포체티노', '영국', '잉글랜드', '아시안컵', '맨유', '가수']
pattern = '|'.join(delete_keywords)
# 키워드 포함 행 제거
filtered_news_df = news_df[~news_df['Title'].str.contains(pattern, case=False, na=False)]
filtered_news_df
```

```python
import nltk
from nltk.corpus import stopwords
import re
import string

# 텍스트 전처리 함수
def text_preprocessor(text, del_number=True, del_bracket_content=True, del_punctuation=True):
    if del_number:
        text = re.sub(r'\d+', '', text)  # 숫자 제거
    if del_bracket_content:
        text = re.sub(r'\[.*?\]|\(.*?\)|\{.*?\}', '', text)  # 괄호와 그 안의 내용 제거
    if del_punctuation:
        text = re.sub(r'[^\w\s]', '', text)  # 문장 부호 제거 (단어와 공백을 제외한 모든 문자)
    return text
        
    # 문장부호 제거
    translator = str.maketrans('', '', string.punctuation)
    text_new = text.translate(translator)
    
    # 영어 stopwords 제거
    stop_words_eng = set(stopwords.words('english'))
    text_new = ' '.join([word for word in text_new.split(' ') if word.lower() not in stop_words_eng])
    
    return text_new


texts = filtered_news_df['Title'].tolist()
filtered_texts = [text_preprocessor(text) for text in texts]
filtered_news_df['Title'] = filtered_texts
filtered_news_df
```

```python
from collections import Counter
#빈도수 기반 추출해보기
titles = filtered_news_df['Title'].tolist()
all_text = ' '.join(titles)
tokens = all_text.split()
word_freq = Counter(tokens)
top_words = word_freq.most_common(100)
top_words
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# TF-IDF기반 추출해보기
titles = filtered_news_df['Title'].tolist()
tfidf_vectorizer = TfidfVectorizer()
# TF-IDF 벡터화 및 학습
tfidf_vectors = tfidf_vectorizer.fit_transform(titles)
# 각 단어에 대한 인덱스
feature_names = tfidf_vectorizer.get_feature_names_out()
# TF-IDF 벡터 출력
# 우선 0.1이상으로 설정
for i, title in enumerate(titles):
    print("Title", i+1)
    vector = tfidf_vectors[i].toarray()[0]
    for j, value in enumerate(vector):
        if value > 0.1:
            print(feature_names[j], ":", value)
```
