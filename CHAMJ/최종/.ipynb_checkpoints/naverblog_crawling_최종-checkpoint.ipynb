```python
#최종
import re
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
from emoji import core
import asyncio
import aiohttp
import nest_asyncio

BLOG_DOMAIN = "https://blog.naver.com/"

nest_asyncio.apply()

async def fetch_and_parse_blog_data(_url, data):
    async with aiohttp.ClientSession() as session:
        async with session.get(_url) as response:
            text = await response.text()
            soup = BeautifulSoup(text.replace("\\", ""), "html.parser")
            parse_blog_data(soup, data)

def parse_blog_data(soup, result_data):
    title_and_url_elems = soup.find_all("a", {"class": "api_txt_lines"})
    desc_elems = soup.find_all("div", {"class": "api_txt_lines"})
    date_elems = soup.find_all("span", {"class": "sub_time"})

    for i in range(len(title_and_url_elems)):
        title = title_and_url_elems[i].text
        url = title_and_url_elems[i]['href']
        desc = desc_elems[i].text
        date = date_elems[i].text
        result_data.append({'url': url, 'title': title, 'description': desc, 'date': date})

async def async_requests(url_lst, data):
    await asyncio.gather(*(fetch_and_parse_blog_data(url_lst[i], data) for i in range(len(url_lst))))

def get_total_blog_count(data):
    data = data.replace('\\', '')
    start_index = data.find('"total":"') + 9
    end_index = data[start_index:start_index + 50].find('"') + start_index
    count = int(data[start_index:end_index])
    return count

async def crawl_blogs(keyword, start_page, end_page, start_date, end_date):
    data = []
    url_template = f'https://s.search.naver.com/p/blog/search.naver?where=blog&sm=tab_pge&api_type=1&query={keyword}&rev=44&start={{start}}&dup_remove=1&post_blogurl=&post_blogurl_without=&nso=so:dd,p:from{end_date}to{start_date}&nlu_query=r_category:29+27&dkey=0&source_query=&nx_search_query={keyword}&spq=0&_callback=viewMoreContents'

    await async_requests([
        url_template.format(start=i * 30)
        for i in range(start_page, end_page)
    ], data)

    return data

def remove_emojis(s):
    return core.replace_emoji(s, replace="")

def extract_blog_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    content = (
        soup.find('div', 'se-main-container') or
        soup.select_one('#postViewArea') or
        soup.select_one('div.se_component_wrap.sect_dsc.__se_component_area > div.se_component.se_paragraph.default > div > div > div > div > div > p')
    )

    if content:
        content_text = content.get_text(strip=True)
        content_text = remove_emojis(content_text)
        url_pattern = r"(http|https)?:\/\/[a-zA-Z0-9-\.]+\.[a-z]{2,}(\S*)?|www\.[a-zA-Z0-9-]+\.[a-z]{2,}(\S*)?|[a-zA-Z0-9-]+\.[a-z]{2,}(\S*)?|([a-zA-Z0-9-]+\.)?naver.com(\S*)?"
        content_text = re.sub(url_pattern, "", content_text)
        content_text = re.sub(r"\u200C|\u200b", "", content_text)
        content_text = ' '.join(content_text.split())
        return content_text
    return None

async def main():
    try:
        # 첫 번째 크롤링: naver_blog_data.csv 파일 생성
        keyword = '고령화'
        start_page = 1
        end_page = 3
        start_date = 20180101
        end_date = 20230901

        crawled_data = await crawl_blogs(keyword, start_page, end_page, start_date, end_date)

        # CSV 파일에 저장
        with open('naver_blog_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=['url', 'title', 'description', 'date'])
            writer.writeheader()
            writer.writerows(crawled_data)

        data_file = 'naver_blog_data.csv'
        # 새로운 CSV 파일
        output_file = 'output.csv'
        # 결과를 저장할 새로운 파일
        merged_output_file = 'merged_output.csv'

        # CSV 파일 불러오기
        df = pd.read_csv(data_file)

        # 새로운 CSV 파일에 본문 내용 추가
        with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['url', 'content_text'])

            for idx, row in df.iterrows():
                url = row['url']
                if not url.startswith(BLOG_DOMAIN):
                    continue

                frameaddr = BLOG_DOMAIN + BeautifulSoup(requests.get(url).text, 'html.parser').find('iframe', id="mainFrame")['src']
                print(frameaddr)

                content_text = extract_blog_content(frameaddr)
                if content_text:
                    print(content_text)
                    print()

                    writer.writerow([url, content_text])

        # 두 CSV 파일 병합 (기존 코드 수정)
        df_output = pd.read_csv(output_file)
        merged_df = df.merge(df_output, on='url', how='inner')

        # 병합된 데이터프레임을 새로운 CSV 파일로 저장
        merged_df.to_csv(merged_output_file, index=False, encoding='utf-8')

    except requests.exceptions.RequestException as er:
        print(f"An error occurred during the request: {er}")
    except Exception as er:
        print(f"An error occurred: {er}")

if __name__ == "__main__":
    asyncio.run(main()) 
```
