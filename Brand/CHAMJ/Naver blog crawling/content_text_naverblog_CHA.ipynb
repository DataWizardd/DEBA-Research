

## remove_emojis 함수: 문자열에서 이모지를 제거

## extract_blog_content 함수: 블로그 글의 URL을 받아 해당 글의 본문을 추출하는 함수

```python
import re
import csv
import requests
from bs4 import BeautifulSoup
from emoji import core
import pandas as pd

BLOG_DOMAIN = "https://blog.naver.com/"

def remove_emojis(s):
    return core.replace_emoji(s, replace="")

def extract_blog_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    content = soup.find('div', 'se-main-container')
    if content is None:
        content = soup.select_one('#postViewArea')
    if content is None:
        content = soup.select_one('div.se_component_wrap.sect_dsc.__se_component_area > div.se_component.se_paragraph.default > div > div > div > div > div > p')

    if content:
        content_text = content.get_text(strip=True)
        content_text = remove_emojis(content_text)
        url_pattern = r"(http|https)?:\/\/[a-zA-Z0-9-\.]+\.[a-z]{2,}(\S*)?|www\.[a-zA-Z0-9-]+\.[a-z]{2,}(\S*)?|[a-zA-Z0-9-]+\.[a-z]{2,}(\S*)?|([a-zA-Z0-9-]+\.)?naver.com(\S*)?"
        content_text = re.sub(url_pattern, "", content_text)
        content_text = re.sub(r"\u200C|\u200b", "", content_text)
        content_text = ' '.join(content_text.split())
        return content_text
    return None

def main():
    try:
        # 불러온 데이터 파일
        data_file = 'naver_blog_data.csv'
        # 새로운 CSV 파일
        output_file = 'output.csv'

        # CSV 파일 불러오기
        df = pd.read_csv(data_file)

        with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['idx', 'content_text'])

            for idx, row in df.iterrows():
                url = row['url']
                if not url.startswith(BLOG_DOMAIN):
                    continue

                frameaddr = BLOG_DOMAIN + BeautifulSoup(requests.get(url).text, 'html.parser').find('iframe', id="mainFrame")['src']
                print(frameaddr)

                content_text = extract_blog_content(frameaddr)
                if content_text:
                    print(content_text)
                    print()

                    writer.writerow([idx, content_text])

    except requests.exceptions.RequestException as er:
        print(f"An error occurred during the request: {er}")
    except Exception as er:
        print(f"An error occurred: {er}")

if __name__ == "__main__":
    main()
```
