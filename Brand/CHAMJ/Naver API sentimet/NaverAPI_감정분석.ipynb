```python
from tqdm import tqdm as tqdm_iterator
```

```python
### 하위의 모든 데이터 결합하여 출력
def get_data_from_path(folder_location, folder_name=False, concat_axis='row'):
    # path_folder 하위의 모든 폴더위치와 내부 file 출력
    df = pd.DataFrame()
    print('Getting data from', len(os.listdir(folder_location)), 'folders...')
    for (path, dir, files) in os.walk(folder_location):
        for file in tqdm_iterator(files):
            path_file = os.path.join(path, file)

            ## 데이터 로딩
            if path_file.split('.')[1] == 'xlsx':
                df_sub = pd.read_excel(path_file)
            elif path_file.split('.')[1] == 'csv':
                df_sub = pd.read_csv(path_file)
                
            ## 키워드 태깅 여부
            if folder_name:
                df_sub['Folder_Name'] = os.path.basename(path)
            
            ## 정리
            if concat_axis == 'col':
                df = pd.concat([df, df_sub], axis=1)
            elif concat_axis == 'row':
                df = pd.concat([df, df_sub], axis=0)
                
    return df

# 데이터로딩
df_news = get_data_from_path(FOLDER_LOCATION, folder_name=FOLDER_NAME)
df = df_news.copy()
## 날짜 인식
df['일자'] = pd.to_datetime(df['일자'].astype(str))
## 중복 처리
df.drop_duplicates(inplace=True, ignore_index=True)
## 카테고리 필터
category_filter = [each for each in df['통합 분류1'].unique() if each.split('>')[0] in CATEGORY]
print(category_filter)
df = df[df['통합 분류1'].apply(lambda x: x in category_filter)].reset_index().iloc[:,1:]
## 전처리
df['제목'] = df['제목'].apply(lambda x: text_preprocessor(x, del_bracket_content=False))
df.shape
```

# 그룹화

```python
# 엑셀 파일에서 데이터 불러오기
df = pd.read_excel('sentiment_analysis_results.xlsx')

# 'sentences' 열의 JSON 문자열을 파이썬 딕셔너리로 변환
df['sentences'] = df['sentences'].apply(lambda x: eval(x))

# 'sentiment' 값을 딕셔너리에서 추출
df['sentiment'] = df['sentences'].apply(lambda x: x[0]['sentiment'])

# '일자' 열을 datetime 형식으로 변환
df['일자'] = pd.to_datetime(df['일자'], format='%Y%m%d')

# 연도와 월에 따라 데이터 그룹화
yearly_grouped = df.groupby([df['일자'].dt.to_period('Y'), df['일자'].dt.to_period('M'), 'sentiment']).size().unstack(fill_value=0)

# 연도별 그룹화된 데이터를 새 엑셀 파일로 저장
yearly_grouped.to_excel("yearly_sentiment_grouped_results.xlsx")
```

## 시계열 데이터

```python
import matplotlib.pyplot as plt

# '일자' 열을 기준으로 'sentiment' 열을 그룹화하여 각 날짜별로 sentiment의 개수를 세기
sentiment_counts = df['sentiment'].groupby(df.index).value_counts().unstack().fillna(0)

# 각 sentiment에 대한 누적 합계 계산
cumulative_counts = sentiment_counts.cumsum()

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(cumulative_counts.index, cumulative_counts['negative'], label='Negative', marker='o')
plt.plot(cumulative_counts.index, cumulative_counts['neutral'], label='Neutral', marker='o')
plt.plot(cumulative_counts.index, cumulative_counts['positive'], label='Positive', marker='o')

plt.title('Sentiment Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Count')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# '일자' 열을 기준으로 'sentiment' 열을 그룹화하여 각 날짜별로 sentiment의 개수를 세기
sentiment_counts = df['sentiment'].groupby(df.index).value_counts().unstack().fillna(0)
# 각 sentiment에 대한 누적 합계 계산
cumulative_counts = sentiment_counts.cumsum()

# 감정을 수치로 매핑
sentiment_mapping = {'positive': 1, 'neutral': 0, 'negative': -1}
df['sentiment'] = df['sentiment'].map(sentiment_mapping)

# 선 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['sentiment'], label='Sentiment', marker='o')
plt.title('Sentiment Over Time')
plt.xlabel('Date')
plt.ylabel('Sentiment')
plt.legend()
plt.grid(True)

# 7일 이동평균 계산
rolling_mean = df['sentiment'].rolling(window=7).mean()

# 이동평균 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(rolling_mean.index, rolling_mean, label='7-Day Moving Average', color='red')
plt.title('7-Day Moving Average of Sentiment')
plt.xlabel('Date')
plt.ylabel('Sentiment')
plt.legend()
plt.grid(True)

plt.show()
```
