```python
# 실제 리뷰 텍스트가 담겨있는 목록만 스크래핑 해오겠다.
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome()
driver.maximize_window()
df = pd.read_csv('draft.csv') #링크가 있는 파일


df_copy = df.copy()

#리뷰데이터를담을 리스트
rev_list = []

for i in range(len(df_copy)):
    driver.get(df_copy['naver_map_url'][i])
    time.sleep(3)
    print('현재 수집중인 장소 : ', df_copy['Name'][i])
    
#최신 순으로 나열한대로
    try:
        reviewnum = driver.find_element(By.CSS_SELECTOR, '#app-root > div > div > div > div:nth-child(7) > div:nth-child(3) > div.place_section.k5tcc > h2 > span.place_section_count').text
        print('실 리뷰 수 : ', reviewnum)

        # 데이터프레임에 리뷰수 추가
        df_copy.loc[i, '리뷰 수'] = reviewnum
    except:
        df_copy.loc[i, '리뷰 수'] = None

# csv파일 저장
df_copy.to_csv('reviews_num.csv', index=False, encoding='utf-8')
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException


for i in range(len(df)):
    try:
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')

        review_lists = soup.select('div.ZZ4OK > a > span')


        if len(review_lists) > 0:
            try:
                WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'div.ZZ4OK > a > span.rvCSr > svg'))).click()
                time.sleep(1)
            except:
                pass

            # 유저리뷰 불러오기
            user_reviews = soup.select('div.ZZ4OK > a > span')

            # \'rev_list'에 추가
            for review in user_reviews:
                rev_list.append([
                    df['Name'][i],
                    '',
                    review.text
                ])
            time.sleep(1)

        else:
            print('리뷰가 존재하지 않음')
            # 'rev_list'에 데이터 추가
            rev_list.append([
                df['Name'][i],
                '',
            ])
            time.sleep(2)

    except NoSuchElementException:
        print('리뷰 텍스트가 인식되지 않음')
        continue
        
driver.close()  
```
